```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

```

```python
from pylab import *
```

```python
x = np.linspace(0, 5, 10)
y = x ** 3

figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('title')
show()

```

```python
subplot(1,2,1)
plot(x, y, 'g*')
subplot(1,2,2)
plot(y, x, 'g*-');
```

```python

dt = 0.01
Fs = 1/dt
t = np.arange(0, 10, dt)
nse = np.random.randn(len(t))
r = np.exp(-t/0.05)

cnse = np.convolve(nse, r)*dt
cnse = cnse[:len(t)]
s = 0.1*np.sin(2*np.pi*t) + cnse

plt.subplot(3, 2, 1)
plt.plot(t, s, 'g-')
```

```python
import random

def d6():
    return random.randint(1,6)

for x in range(100):
    print(str(d6()))
        
```

```python
def rng(seed=1):
    def rand():
        nonlocal seed
        seed = (7 * seed) % 101
        return (seed - 1) % 10 + 1
    
    return rand

rand = rng(11)

for x in range(10):
    print(rand())
```

```python
nums = [rand() for x in range(1000)]
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.charts import Histogram
```

```python
hist = Histogram(nums, bins=10, tools=False,
                 title="rng", ylabel="frequency", xlabel="distributions")
```

```python
show(hist)
```
